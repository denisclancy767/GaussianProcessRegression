```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import (RBF, Matern, RationalQuadratic,
                                              ExpSineSquared, DotProduct,
                                              ConstantKernel)
```

```python
merged = pd.read_csv('/content/gdrive/MyDrive/merged.csv')
```

```python
merged = merged[merged['Energy (kWh)'] !=0]
merged = merged[merged['Total Units'] !=0]
```

```python
merged = merged.reset_index()
merged
```

```python
kernelrbf = 1 * RBF(length_scale=3.0, length_scale_bounds=(1e-2, 1e2))
kernelexpsinesqrd = 1 * ExpSineSquared(length_scale=3.0, length_scale_bounds=(1e-2, 1e2))
kernelrationalquadratic = 1.0 * RationalQuadratic(length_scale=1.0, alpha=0.1)
kernelconstant = ConstantKernel(0.1, (0.01, 10.0))
kernelmatern = 1.0 * Matern(length_scale=1.0, length_scale_bounds=(1e-1, 10.0),nu=1.5)

gprrbf = GaussianProcessRegressor(kernel=kernelrbf, n_restarts_optimizer=100, normalize_y = True,alpha = 1e-5)
gprexpsinesqrd = GaussianProcessRegressor(kernel=kernelexpsinesqrd, n_restarts_optimizer=100, normalize_y = True,alpha = 1e-5)
gprrationalquadratic = GaussianProcessRegressor(kernel=kernelrationalquadratic, n_restarts_optimizer=100, normalize_y = True,alpha = 1e-5)
gprconstant = GaussianProcessRegressor(kernel=kernelconstant, n_restarts_optimizer=100, normalize_y = True,alpha = 1e-5)
gprmatern = GaussianProcessRegressor(kernel=kernelmatern, n_restarts_optimizer=100, normalize_y = True,alpha = 1e-5)
```

```python
stop = merged.iloc[:]['Stop']
stop = np.array(stop).reshape(-1,1)
energy = merged.iloc[:]['Energy (kWh)']
energy = np.array(energy).reshape(-1,1)
total = merged.iloc[:]['Total Units']
total = np.array(total).reshape(-1,1)
success = merged.iloc[:]['Parts successful']
success = np.array(success).reshape(-1,1)
scrap = merged.iloc[:]['Parts scrapped']
scrap = np.array(scrap).reshape(-1,1)
```

```python
energynormed = (energy - np.mean(energy, axis=0))/(energy.std())

totalnormed = (total - np.mean(total, axis=0))/(total.std())
```

```python
rng = np.random.RandomState(1)
training_indices = rng.choice(np.arange(energy.size), size=6, replace=False)
X_train, y_train = np.array(total[training_indices]).reshape(-1,1), np.array(energy[training_indices]).reshape(-1,1)
```

```python
gprrbf.fit(total,energy)
gprexpsinesqrd.fit(total,energy)
gprrationalquadratic.fit(total,energy)
gprconstant.fit(total,energy)
gprmatern.fit(total,energy)
```

```python
plt.plot(totalnormed,energynormed,'o')
#plt.plot(X_train,y_train)
```

```python
X = np.linspace(start=0, stop=150, num=1000).reshape(-1,1)
predict = gprrbf.predict(X)
plt.plot(X,predict)
```

```python
X = np.linspace(start=0, stop=150, num=1000).reshape(-1,1)
predict = gprrbf.predict(X)

mean_prediction, std_prediction = gprrbf.predict(X, return_std=True)

plt.plot(X, mean_prediction, label="Mean prediction",color='blue')
plt.plot(X,predict)
plt.fill_between(
    X.ravel(),
    mean_prediction - 1.96 * std_prediction,
    mean_prediction + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)
plt.scatter(total,energy,s= 0.8,color='magenta',label='Observations')
plt.legend()
```
